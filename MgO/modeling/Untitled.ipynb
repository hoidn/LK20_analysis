```python
import scipy
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
from scipy.interpolate import interp1d
```

```python
def pulse_profile(width):
    #x = np.arange(-width, width, 1.)
    x = np.arange(0, 2 * width, width + 1.)
    full_gaussian = gaussian(x, width, width)
    interpolator = interp1d(x - width, full_gaussian, fill_value = 0.)
    return interpolator
```

```python
p = pulse_profile(5)
```
